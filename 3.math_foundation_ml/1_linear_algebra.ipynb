## Understanding np.set_printoptions(suppress=True)

**suppress=True** prevents NumPy from using scientific notation (like 1e-5) when printing arrays. Instead, it displays numbers in regular decimal format.

This makes output more readable when dealing with very small or very large numbers.

```python
# Example 1: WITHOUT suppress=True (default behavior)
np.set_printoptions(suppress=False)
arr = np.array([0.00001, 0.0002, 0.003, 123456.789])
print("Without suppress=True:")
print(arr)
```

```python
# Example 2: WITH suppress=True
np.set_printoptions(suppress=True)
arr = np.array([0.00001, 0.0002, 0.003, 123456.789])
print("With suppress=True:")
print(arr)
```

```python
# Example 3: Practical use case - Matrix with small decimal values
matrix = np.array([[0.0001, 0.0002], [0.00003, 0.00004]])
print("Matrix with suppress=True (cleaner for analysis):")
print(matrix)
```

**Eculidean Distance**

**Definition:** Euclidean distance is the straight-line distance between two points in space. It's the most common distance metric in mathematics and machine learning.

**Formula:** For two vectors v1 and v2:

$$d = \sqrt{(v1[0] - v2[0])^2 + (v1[1] - v2[1])^2 + ... + (v1[n] - v2[n])^2}$$

Or equivalently: the square root of the sum of squared differences.

**Why it matters:**
- Used to measure similarity between data points
- Foundation for clustering algorithms (K-means, KNN)
- Determines "closeness" in recommendation systems and classification
- Smaller distance = more similar vectors

```python
#Eculidean distance
euclidean_distance = np.linalg.norm(v1 - v2)
print("Euclidean distance between v1 and v2:", euclidean_distance)
```

**Manhattan Distance**

**What it is:** Manhattan distance is the sum of the absolute differences between corresponding coordinates of two vectors. It is called "Manhattan" distance because it measures distance the way you would move along city blocks, not in a straight line.

**Formula:** For two vectors v1 and v2:

$$d = |v1[0] - v2[0]| + |v1[1] - v2[1]| + ... + |v1[n] - v2[n]|$$

**Use cases:**
- Useful when movement happens in grids or along axes, such as city routing
- Often used in machine learning when you want a distance metric that is less sensitive to large outliers than Euclidean distance
- Common in clustering, nearest-neighbor methods, and feature comparison
- Helpful when features are sparse or when diagonal distance is not meaningful

**Example:** For v1 = [1, 2, 3] and v2 = [4, 5, 6], the Manhattan distance is |1-4| + |2-5| + |3-6| = 9

```python
#Manhattan distance
manhattan_distance = np.sum(np.abs(v1 - v2))
print("Manhattan distance between v1 and v2:", manhattan_distance)
```

**Weights and Bias (Short Note)**

- **Weight (`w`)**: scales an input feature and determines its importance. In linear models it controls the slope.
- **Bias (`b`)**: an added constant (intercept) that shifts the output; allows non-zero output when inputs are zero.

Model form (scalar / single-feature):

$$y = w\cdot x + b$$

Shapes (examples):
- Single feature: `w` is scalar, `b` is scalar
- Multiple features: `W` shape = `(output_dim, input_dim)`, `b` shape = `(output_dim,)`

Where & why:
- Used in linear regression, logistic regression, dense/conv layers in neural nets.
- `W` learns feature importance; `b` shifts model output so it's not forced through origin.
- Training updates `W` and `b` via gradients to minimize the loss.

```python
# Diagram: plot y = w*x + b and annotate weight (slope) and bias (intercept)
import matplotlib.pyplot as plt
w = 2.0
b = 0.5
x = np.linspace(-2, 5, 100)
y = w * x + b

plt.figure(figsize=(6,4))
plt.plot(x, y, label=f"y={w}x+{b}")
# annotate bias at x=0
plt.scatter([0], [b], color='red')
plt.annotate('bias (b)', xy=(0, b), xytext=(1, b+1), arrowprops=dict(arrowstyle='->'))
# annotate weight (slope) between x=0 and x=1
x1, x2 = 0, 1
y1, y2 = w*x1 + b, w*x2 + b
plt.plot([x1, x2], [y1, y2], color='green', linewidth=3, alpha=0.6)
plt.annotate('weight (w) = slope', xy=((x1+x2)/2, (y1+y2)/2), xytext=(2, (y1+y2)/2 - 2), arrowprops=dict(arrowstyle='->'))

plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear model: y = w*x + b')
plt.legend()
plt.tight_layout()
plt.show()
```

**Gradient Descent Demo — Learning `w` and `b`**

We'll fit a simple linear model `y = w*x + b` to synthetic data using mean-squared error and gradient descent. The demo shows parameter updates, loss over iterations, and the fitted line compared to data.

```python
# Gradient descent demo for y = w*x + b
import numpy as np
import matplotlib.pyplot as plt

# Synthetic data: y = 2.0*x + 1.0 + noise
rng = np.random.default_rng(0)
x = rng.uniform(-3, 3, size=50)
y_true_w, y_true_b = 2.0, 1.0
y = y_true_w * x + y_true_b + rng.normal(scale=0.8, size=x.shape)

# Initialize parameters
w = 0.0
b = 0.0
lr = 0.01
n_epochs = 200
n = x.size

loss_history = []
params_history = []

for epoch in range(n_epochs):
    # predictions
    y_pred = w * x + b
    # loss (MSE)
    loss = np.mean((y_pred - y)**2)
    loss_history.append(loss)
    params_history.append((w, b))
    # gradients
    dw = (2/n) * np.sum((y_pred - y) * x)
    db = (2/n) * np.sum(y_pred - y)
    # update
    w -= lr * dw
    b -= lr * db

# final predictions
y_pred_final = w * x + b
print(f"Learned w={w:.4f}, b={b:.4f} (true w={y_true_w}, b={y_true_b})")

# Plot data and fitted line
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.scatter(x, y, label='data')
xs = np.linspace(x.min()-0.5, x.max()+0.5, 100)
plt.plot(xs, w*xs + b, color='red', label=f'fitted: y={w:.2f}x+{b:.2f}')
plt.plot(xs, y_true_w*xs + y_true_b, color='green', linestyle='--', label='true')
plt.legend()
plt.title('Data and fitted line')
plt.xlabel('x')
plt.ylabel('y')

# Plot loss
plt.subplot(1,2,2)
plt.plot(loss_history)
plt.title('MSE Loss over iterations')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.tight_layout()
plt.show()
```
