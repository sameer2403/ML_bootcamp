## Understanding np.set_printoptions(suppress=True)

**suppress=True** prevents NumPy from using scientific notation (like 1e-5) when printing arrays. Instead, it displays numbers in regular decimal format.

This makes output more readable when dealing with very small or very large numbers.

```python
# Example 1: WITHOUT suppress=True (default behavior)
np.set_printoptions(suppress=False)
arr = np.array([0.00001, 0.0002, 0.003, 123456.789])
print("Without suppress=True:")
print(arr)
```

```python
# Example 2: WITH suppress=True
np.set_printoptions(suppress=True)
arr = np.array([0.00001, 0.0002, 0.003, 123456.789])
print("With suppress=True:")
print(arr)
```

```python
# Example 3: Practical use case - Matrix with small decimal values
matrix = np.array([[0.0001, 0.0002], [0.00003, 0.00004]])
print("Matrix with suppress=True (cleaner for analysis):")
print(matrix)
```

## 1. scalar, vector, Matrix & Tensor in NumPy

```python
#scalar -single value, not an array
scalar = 5.0
print("Scalar value:", scalar)
print("Type of scalar:", type(scalar))
```

```python
# vector - 1D array
vector = np.array([1, 2, 3])
print("Vector:", vector)
print("Type of vector:", type(vector))
print("Vector shape:", vector.shape)
```

```python
# matrix - 2D array
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix:\n", matrix)
print("Type of matrix:", type(matrix))
print("Matrix shape:", matrix.shape)
```

```python
# Tensor - multi-dimensional array (3D or higher)
tensor = np.array(
                [
                    [
                        [1, 2], [3, 4]
                    ], 
                    [
                        [5, 6], [7, 8]
                    ]
                ]
    )
print("Tensor:\n", tensor)
print("Type of tensor:", type(tensor))
print("Tensor shape:", tensor.shape)
```

**2. Creating vectors and matrices for ML style data**

```python
# Imagine we have 3 samples and 2 features
# Feature 1: hours studied
# Feature 2: number of practice questions
# target: pass (1)/ fail (0)
X = np.array(
    [
        [5, 50], 
        [3, 30], 
        [8, 80]
    ]
) # shape (3, 2) # 3 samples, 2 features

Y = np.array([1, 0, 1]) # shape (3,)

print("Feature matrix X:\n", X)
print("-" * 30)
print("Target vector Y:\n", Y)
print("-" * 30)
x1= X[:, 0] # hours studied
x2 = X[:, 1] # number of practice questions
print("Feature 1 (hours studied):", x1)
print("Feature 2 (number of practice questions):", x2)
print("-" * 30 )
x3 =X[0, :] # first sample (5 hours, 50 questions)
print("First sample (5 hours, 50 questions):", x3)
print(x3.shape) # shape (2,) - 1D array with 2 elements

```

**3. Basic operations: addition, scalar multiplication, transpose**

```python
# two vectors with 3 elements each
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
print("Vector 1:", v1)
print("Vector 2:", v2)
print("-" * 30)
#vector addition
v_add = v1 + v2
print("Vector addition (v1 + v2):", v_add)
print("Type of result:", type(v_add))
print("Shape of result:", v_add.shape)
print("-" * 30)

# scalar multiplication
scalar = 2
v_scalar_mult = scalar * v1
print(f"Scalar multiplication ({scalar} * v1):", v_scalar_mult)
print("Type of result:", type(v_scalar_mult))

print("-" * 30)

# dot product
dot_product = np.dot(v1, v2)
print("Dot product (v1 . v2):", dot_product)
print("Type of result:", type(dot_product)) 
print("-" * 30)

# matrix multiplication
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])
print("Matrix 1:\n", m1)
print("Matrix 2:\n", m2)
print("-" * 30)
m_mult = np.dot(m1, m2)
print("Matrix multiplication (m1 . m2):\n", m_mult)
print("Type of result:", type(m_mult))
print("Shape of result:", m_mult.shape)
print("-" * 30)
#scalar multiplication
scalar = 0.5
m_scalar_mult = scalar * m1
print(f"Scalar multiplication ({scalar} * m1):\n", m_scalar_mult)
print("-" * 30)

#matrix addition
m_add = m1 + m2
print("Matrix addition (m1 + m2):\n", m_add)
print("Type of result:", type(m_add))
print("Shape of result:", m_add.shape)
print("-" * 30)

#transpose of a matrix
m_transpose = m1.T
print("Transpose of m1:\n", m_transpose)
print("Shape of transpose:", m_transpose.shape)



```
