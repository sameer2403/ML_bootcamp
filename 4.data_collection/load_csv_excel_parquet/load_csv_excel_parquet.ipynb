**Compare file sizes**

```python
# define a function to compare file sizes
def compare_file_sizes(path):
    size_bytes = os.path.getsize(path)
    size_kb = round(size_bytes / 1024, 2)
    size_mb = round(size_kb / 1024, 2)
    return size_bytes, size_kb, size_mb

for file in ["titanic.csv", "titanic.xlsx", "titanic.parquet"]:
    size_bytes, size_kb, size_mb = compare_file_sizes(file)
    print(f"{file}: {size_bytes} bytes, {size_kb} KB, {size_mb} MB")
```
