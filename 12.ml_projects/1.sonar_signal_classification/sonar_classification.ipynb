**Dataset description:**

The Sonar dataset contains sonar signal readings collected by bouncing sound waves off objects underwater, with the goal of predicting whether the object is a rock or a mine. Each row represents one sonar return, described using numerical energy measurements from multiple frequencies.

The dataset has 60 numerical feature columns that represent the strength of sonar signals at different frequencies. The final column is the target label, where R denotes rock and M denotes mine.

**4. Data Preprocessing**

```python
# seperate feature & tgt
X = df.drop(columns=60)
y=df[60]
```

```python
X[:5]
```

```python
y[:5]
```

```python
# train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.2,
    stratify=y, # Ensures the class proportions in y are preserved in both the train and test splits.
    random_state= RANDOM_STATE # Seeds the random number generator so the split is reproducible — same rows go to train/test every time you rerun the code.
    
)
```

```python
print("Dataset Shape:",X.shape)
print("Training Dataset Shape:",X_train.shape)
print("Test Dataset Shape:",X_test.shape)
```

```python
#feature scaling
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**5.Model training**

**Support Vector Classifier (SVC) works well for high dimensional data**

```python
model = SVC()
```

```python
# train the model
model.fit(X_train_scaled, y_train)
```

**6.Model Evaluation**

```python

y_train_pred = model.predict(X_train_scaled)
train_acc= accuracy_score(y_train, y_train_pred)
print(f"Training Data Accurcay:{round(train_acc*100,2)}")
```

```python
# test accuracy_score
y_test_pred = model.predict(X_test_scaled)
test_acc= accuracy_score(y_test, y_test_pred)
print(f"Test Data Accuracy:{round(test_acc*100,2)}")
```

**Classification report**

```python
print("Training data-classification_report")
print(classification_report(y_train,y_train_pred))
```

```python
print("Test data-classification_report")
print(classification_report(y_test,y_test_pred))
```

**7.Building a Predictive System**

```python
def predict_object(input_features):
    # scale features
    scaled_features = scaler.transform([input_features])
    # get prediction from the model
    prediction= model.predict(scaled_features)
    print("Model prediction:",prediction)
    if prediction[0] == 1:
        print("Object identified as Mine")
    else:
        print("Object identified as Rock")     
    
```
