**6.1 Train Test Split**

**Steps:**
1. Load the Titanic data
2. Separate features and target
3. Split into train and test sets

---

**Data Leakage**

Data leakage happens when your model accidentally learns information during training that it would never see in real life when making predictions (eg: test data / real world scenario).
Because of this, the model looks very accurate during training but performs poorly on new data.

**Simple example:**

Imagine predicting house prices, and you accidentally include a column called “final_sold_price” while training. The model uses that to predict itself. It looks perfect in training but fails in real world.

**One line summary:**

Data leakage is when the model cheats by seeing future or hidden information during training.

**How to prevent it:**
- Split data first before preprocessing.
- Do scaling and feature engineering only based on training data.
- Remove columns that contain future or target related information.

**Golden Rule:**
- always splt your data first
- never learn patterns from data first

```python
import pandas as pd
from sklearn.model_selection import train_test_split

pd.set_option("display.max_columns", None)
```

**2. Load the dataset**

```python
df = pd.read_csv("titanic_dataset.csv")
print("Shape of the dataset:", df.shape)
print("Columns in the dataset:", df.columns.tolist())
df.head()
```

```python
# dropping "alive" column as it is redundant due to the presence of "survived" column
df=df.drop(columns=["alive"])
```

```python
print("Shape of the dataset:", df.shape)
print("Columns in the dataset:", df.columns.tolist())
df.head()
```

**3. Drop rows with missing target (safe before train test split)**

In this dataset, target column (survived) doesn't have any missing values)

```python
df.isnull().sum()
```

```python
target = "survived"
```

```python
missing_target_count = df[target].isna().sum()
print("Rows with missing target:", missing_target_count)

# drop the rows that has missing target value
df = df.dropna(subset=[target])

print("Shape after dropping missing target rows", df.shape)
```

**4. Separate X (features) and y (target)**

```python
X = df.drop(columns=[target])
y = df[target]
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
X[:5]
```

```python
y[:5]
```

```python
print("Target distribution on full data:")
print(y.value_counts())

print("-"*50)

print("Target distribution in percentage")
print(y.value_counts(normalize=True).round(3))
```

**5. Train Test Split**

- Train set -> ML Model Training
- Test set  -> ML Model Evaluation

X (features)  -> training_features (X_train); test_features (X_test)

y (target)    -> training_targets (y_train); test_targets (y_test)

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y, 
    test_size=0.2,  # 20% data for testing, 80% for training
    random_state=42, # for reproducibility: always get the same split when you run the code multiple times
    stratify=y,   # even distribution of target in train and test sets
    )
```

```python
print("Shapes:")
print("X:", X.shape)
print("y:", y.shape)
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

**Training process:**

X_train, y_train -> ML model learns the pattern -> trained ML model


**Evaluation Process**

X_test -> trained ML model -> y_pred (prediction from the trained model)

Compare y_pred & y_test (true labels)

```python
print("Target distribution on full data:")
print(y.value_counts())

print("Target distribution in percentage")
print(y.value_counts(normalize=True).round(3))

print("-"*50)

print("Target distribution on training data:")
print(y_train.value_counts())

print("Target distribution in percentage")
print(y_train.value_counts(normalize=True).round(3))

print("-"*50)

print("Target distribution on test data:")
print(y_test.value_counts())

print("Target distribution in percentage")
print(y_test.value_counts(normalize=True).round(3))

print("-"*50)
```

**6.Save splits**

```python
train_df = X_train.copy()
train_df[target] = y_train
test_df = X_test.copy()
test_df[target] = y_test
```
